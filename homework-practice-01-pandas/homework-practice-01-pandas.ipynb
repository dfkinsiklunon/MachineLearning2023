#### 5. [0.25 баллов] Какие 5 товаров были самыми дешёвыми и самыми дорогими? (по item_name)

Для этого будет удобно избавиться от дубликатов и отсортировать товары. Не забудьте про количество товара.

```python
# Загрузите данные из файла 'chipotle.tsv'
df = pd.read_table('chipotle.tsv')

# Преобразуйте столбец 'item_price' в числовой формат
def convert_price_to_float(price):
    return float(price.replace('$', '').strip())

df['item_price'] = df['item_price'].apply(convert_price_to_float)

# Создайте столбец 'total_price', учитывающий общую стоимость каждого продукта (цена * количество)
df['total_price'] = df['item_price'] * df['quantity']

# Удалите дубликаты продуктов
unique_products = df.drop_duplicates(subset=['item_name'])

# Сортируйте DataFrame по столбцу 'total_price' (по убыванию)
sorted_df = unique_products.sort_values(by='total_price', ascending=True)

# Выберите 5 самых дешевых и 5 самых дорогих продуктов
cheapest_products = sorted_df.head(5)
most_expensive_products = sorted_df.tail(5)

# Выведите результаты
print("5 самых дешевых продуктов:")
print(cheapest_products[['item_name', 'total_price']])

print("\n5 самых дорогих продуктов:")
print(most_expensive_products[['item_name', 'total_price']])
```

#### 11. [0.75 баллов] В каком количестве заказов есть товар, который стоит более 40% от суммы всего чека?

Возможно, будет удобно посчитать отдельно среднюю стоимость заказа, добавить ее в исходные данные и сделать необходимые проверки.

*Данный комментарий стоит воспринимать как подсказку к одному из вариантов решений задания. Если в вашем варианте решения он не нужнен, это не страшно*

```python
import pandas as pd

# Загрузите DataFrame из файла 'chipotle.tsv'
df = pd.read_csv('chipotle.tsv', sep='\t')

# Преобразуйте столбец 'item_price' в числовой формат
df['item_price'] = df['item_price'].apply(lambda x: float(x[1:]))

# Добавьте столбец 'order_total' с общей суммой каждого заказа
df['order_total'] = df['quantity'] * df['item_price']

# Найдите среднюю стоимость заказа
average_order_total = df.groupby('order_id')['order_total'].sum().mean()

# Инициализируйте счетчик заказов, в которых товар(ы) стоит более 40% от средней стоимости заказа
count_orders_above_40_percent = 0

# Переберите каждый заказ и проверьте условие
for order_id, group in df.groupby('order_id'):
    order_total = group['order_total'].sum()
    if any(group['order_total'] > 0.4 * average_order_total):
        count_orders_above_40_percent += 1

# Выведите количество заказов, удовлетворяющих условию
print(f"Количество заказов с товаром(ами) стоимостью более 40% от средней стоимости заказа: {count_orders_above_40_percent}")
```

#### 12. [0.75 баллов] Предположим, что в данных была ошибка и Diet Coke (choice_description), который стоил $1.25, должен был стоить 1.35. Скорректируйте данные в таблицы и посчитайте, на какой процент больше денег было заработано с этого товара. Не забывайте, что количество товара не всегда равно 1.

```python
import pandas as pd

# Load the DataFrame from the 'chipotle.tsv' file
df = pd.read_csv('chipotle.tsv', sep='\t')

# Make a copy of the DataFrame to avoid modifying the original data
df_copy = df.copy()

# Convert the 'item_price' column to numeric if it's not already
df_copy['item_price'] = pd.to_numeric(df_copy['item_price'], errors='coerce')

# Calculate the total revenue before the change (excluding rows with NaN values)
total_revenue_before_change = df_copy['item_price'].dropna().sum()

# Modify the price of 'Diet Coke' to increase it by 10%
df_copy.loc[df_copy['choice_description'].str.contains('Diet Coke', case=False, na=False), 'item_price'] *= 1.1

# Calculate the total revenue after the change (excluding rows with NaN values)
total_revenue_after_change = df_copy['item_price'].dropna().sum()

# Calculate the percentage increase in revenue
if total_revenue_before_change == 0:
    percentage_increase = 0  # To avoid division by zero if there's no revenue initially
else:
    percentage_increase = ((total_revenue_after_change - total_revenue_before_change) / total_revenue_before_change) * 100

# Print the percentage increase in revenue after the correction
print(f"Percentage increase in revenue after the correction: {percentage_increase:.2f}%")
```

## 2. Визуализации и matplotlib

При работе с данными часто неудобно делать какие-то выводы, если смотреть на таблицу и числа в частности, поэтому важно уметь визуализировать данные. В этом разделе мы этим и займёмся.

У matplotlib, конечно, же есть [документация](https://matplotlib.org/users/index.html) с большим количеством [примеров](https://matplotlib.org/examples/), но для начала достаточно знать про несколько основных типов графиков:
- plot — обычный поточечный график, которым можно изображать кривые или отдельные точки;
- hist — гистограмма, показывающая распределение некоторое величины;
- scatter — график, показывающий взаимосвязь двух величин;
- bar — столбцовый график, показывающий взаимосвязь количественной величины от категориальной.

В этом задании вы попробуете построить каждый из них. Не менее важно усвоить базовые принципы визуализаций:
- на графиках должны быть подписаны оси;
- у визуализации должно быть название;
- если изображено несколько графиков, то необходима поясняющая легенда;
- все линии на графиках должны быть чётко видны (нет похожих цветов или цветов, сливающихся с фоном);
- если отображена величина, имеющая очевидный диапазон значений (например, проценты могут быть от 0 до 100), то желательно масштабировать ось на весь диапазон значений (исключением является случай, когда вам необходимо показать малое отличие, которое незаметно в таких масштабах).
- сетка на графике помогает оценить значения в точках на глаз, это обычно полезно, поэтому лучше ее отрисовывать.

#### 1. [1 балл] Постройте гистограмму распределения сумм покупок и гистограмму средних цен отдельных видов продуктов item_name. 

Изображайте на двух соседних графиках. Для этого может быть полезен subplot.

```python
# Sample dataset
data = {
    'item_name': ['Item A', 'Item B', 'Item A', 'Item C', 'Item B', 'Item C', 'Item A', 'Item B'],
    'purchase_amount': [10.50, 15.75, 8.00, 12.25, 14.50, 10.00, 9.75, 11.00],
    'price_per_unit': [2.50, 3.00, 2.50, 4.25, 3.00, 4.00, 2.50, 3.00]
}

df = pd.DataFrame(data)

# Create a histogram of purchase amounts
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(df['purchase_amount'], bins=5, color='skyblue', edgecolor='black')
plt.title('Distribution of Purchase Amounts')
plt.xlabel('Purchase Amount')
plt.ylabel('Frequency')

# Create a histogram of average prices of individual item types
average_prices = df.groupby('item_name')['price_per_unit'].mean()
plt.subplot(1, 2, 2)
plt.bar(average_prices.index, average_prices.values, color='lightgreen')
plt.title('Average Prices of Individual Item Types')
plt.xlabel('Item Type')
plt.ylabel('Average Price')
plt.xticks(rotation=45)

# Adjust layout for better visualization
plt.tight_layout()

# Show the plots
plt.show()
```

#### 2. [1 балл] Постройте график зависимости суммы покупок от дней.

```python

# Sample dataset with purchase dates and amounts
data = {
    'purchase_date': ['2023-09-01', '2023-09-02', '2023-09-03', '2023-09-04', '2023-09-05'],
    'purchase_amount': [100, 150, 80, 120, 200]
}

df = pd.DataFrame(data)

# Convert the 'purchase_date' column to datetime
df['purchase_date'] = pd.to_datetime(df['purchase_date'])

# Sort the DataFrame by date
df = df.sort_values(by='purchase_date')

# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(df['purchase_date'], df['purchase_amount'], marker='o', linestyle='-')
plt.title('Amount of Purchases Over Days')
plt.xlabel('Date')
plt.ylabel('Purchase Amount')
plt.grid(True)

# Format x-axis labels to show dates nicely
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

#### 3. [1 балл] Постройте средних сумм покупок по дням недели (bar plot).

```python
# Sample dataset with purchase dates and amounts
data = {
    'purchase_date': ['2023-09-01', '2023-09-02', '2023-09-03', '2023-09-04', '2023-09-05'],
    'purchase_amount': [100, 150, 80, 120, 200]
}

df = pd.DataFrame(data)

# Convert the 'purchase_date' column to datetime
df['purchase_date'] = pd.to_datetime(df['purchase_date'])

# Extract the day of the week and calculate average purchase amounts
df['day_of_week'] = df['purchase_date'].dt.day_name()
average_purchase_by_day = df.groupby('day_of_week')['purchase_amount'].mean()

# Define the order of days of the week for sorting
day_order = [
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
]

# Sort the data by the defined order
average_purchase_by_day = average_purchase_by_day.reindex(day_order)

# Create a bar plot
plt.figure(figsize=(10, 6))
average_purchase_by_day.plot(kind='bar', color='skyblue')
plt.title('Average Purchase Amount by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Purchase Amount')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

#### 4. [1 балл] Постройте график зависимости денег за товар от купленного количества (scatter plot).

```python
# Sample dataset with product name, quantity, and total money spent
data = {
    'product_name': ['Product A', 'Product B', 'Product C', 'Product D', 'Product E'],
    'purchase_quantity': [10, 15, 8, 12, 20],
    'total_money_spent': [150, 225, 120, 180, 300]
}

# Create a DataFrame from the sample data
df = pd.DataFrame(data)

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['purchase_quantity'], df['total_money_spent'], c='blue', alpha=0.7, edgecolors='k')
plt.title('Dependence of Money Spent on Purchase Quantity')
plt.xlabel('Purchase Quantity')
plt.ylabel('Total Money Spent')
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
